# Smoothstep general form



```python
from scipy.special import binom
from matplotlib import pyplot as plt
from scipy.special import expit as sigmoid
import numpy as np 

def smoothstep(n):
    x = np.linspace(0, 1, 100)
    sum_array = np.zeros(x.shape)
    
    for k in range(n+1):
        sum_array += binom(n + k, k) * binom(2*n + 1, n - k) * (-x)**k
    
    S = x**(n+1) * sum_array
    return S
```

```python
for i in range(4): 
    S = smoothstep(i)
    plt.plot(S)
```

# Similarity to sigmoid

```python
x = np.linspace(-9, 9, 100)
S = smoothstep(14)
Sig = sigmoid(x)
plt.plot(S)
plt.plot(Sig)
```
